```python
# ============================================================
# Cell 2 — FIX 1: Download folder from Google Drive
# BUG: Original used gdown.download() with a FOLDER id via
#      /uc?id= — only works for single files, ALWAYS fails
#      for folder IDs.
# FIX: Use gdown.download_folder() for folder links.
# ============================================================
import gdown, os

folder_url = 'https://drive.google.com/drive/folders/10eGHnUJ0c0hZlZ5O7gm3f190wHzM3rTq?usp=drive_link'
output_dir = 'Kidney-CT-Scan-data'

if not os.path.exists(output_dir):
    # FIX: download_folder() instead of download() for folder links
    gdown.download_folder(
        url=folder_url,
        output=output_dir,
        quiet=False,
        use_cookies=False,
    )
    print(f'Downloaded to: {output_dir}')
else:
    print(f'Already exists: {output_dir}')
```

```python
# ============================================================
# Cell 3 — FIX 2 & 3: Zip download + unzip (alternative)
# FIX 2: fuzzy=True handles sharing/folder-style links.
# FIX 3: Added the unzip step that was completely missing.
# ============================================================
import gdown, zipfile, os

zip_url    = 'https://drive.google.com/drive/folders/10eGHnUJ0c0hZlZ5O7gm3f190wHzM3rTq?usp=drive_link'
output_zip = 'Kidney-CT-Scan-data.zip'

if not os.path.exists(output_zip):
    gdown.download(zip_url, output_zip, quiet=False, fuzzy=True)
else:
    print(f'Zip already exists: {output_zip}')

# FIX 3: Unzip step was completely missing in original
if os.path.exists(output_zip):
    with zipfile.ZipFile(output_zip, 'r') as z:
        z.extractall('Kidney-CT-Scan-data')
    print('Extracted successfully!')
```

```python
# ============================================================
# Cell 4 — SHORTCUT ⚡
# Kidney-CT-Scan-data.zip already exists in research/ folder.
# Just extract it directly — no download needed!
# ============================================================
import zipfile, os

local_zip  = 'Kidney-CT-Scan-data.zip'   # already in research/
extract_to = 'Kidney-CT-Scan-data'

if os.path.exists(local_zip):
    os.makedirs(extract_to, exist_ok=True)
    with zipfile.ZipFile(local_zip, 'r') as z:
        z.extractall(extract_to)
    print(f'✅ Extracted {local_zip} -> {extract_to}')
else:
    print('Zip not found locally — run Cell 2 or 3 to download first.')
```
