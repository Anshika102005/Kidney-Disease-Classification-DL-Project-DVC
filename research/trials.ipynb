```python
import os, zipfile

# Use raw string (r'...') to avoid escape character issues
zip_path = r'C:\Users\LENOVO\Kidney-Disease-Classification-DL-Project-DVC\kidney.zip'

if os.path.exists(zip_path):
    print(f"File size: {os.path.getsize(zip_path)/1024/1024:.2f} MB")
    
    # List contents without extracting
    with zipfile.ZipFile(zip_path, 'r') as z:
        files = z.namelist()
        print(f"\nTotal files: {len(files)}")
        print(f"\nFirst 10 files:")
        for f in files[:10]:
            print(f"  {f}")
else:
    print("ZIP file not found. Check the filename.")
```
