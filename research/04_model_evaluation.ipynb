```python
from dataclasses import dataclass 
from pathlib import Path 

@dataclass(frozen=True)
class EvaluationConfig:
	path_of_model: Path 
	training_data: Path 
	all_params:  dict 
	mlflow_uri: str
	params_image_size: list
	params_batch_size: int
```

```python
from pathlib import Path
from cnnClassifier.constants import * 
from cnnClassifier.utils.common import read_yaml, create_directories, save_json
```

```python
class ConfigurationManager:
    def __init__(
        self,
        config_filepath = CONFIG_FILE_PATH,
        params_filepath = PARAMS_FILE_PATH):
        self.config = read_yaml(config_filepath)
        self.params = read_yaml(params_filepath)
        create_directories([self.config.artifacts_root])

    def get_evaluation_config(self) -> EvaluationConfig:
        eval_config = EvaluationConfig(
            path_of_model=Path("artifacts/training/trained_model.h5"),
            training_data=Path("artifacts/data_ingestion/train"),
            all_params=dict(self.params),
            mlflow_uri="https://dagshub.com/Anshika102005/Kidney-Disease-Classification-DL-Project-DVC.mlflow",
            params_image_size=self.params.IMAGE_SIZE,
            params_batch_size=self.params.BATCH_SIZE
        )
        return eval_config
```
